## Total de Evoluções do quadro clínico

#### 1=Cura
#### 2=Óbito
#### 9= Ignorado

```python
df.groupby('EVOLUCAO').size().sort_values().tail(40).plot(kind='barh', figsize=(8,8))
#1=CURA
#2=ÓBITO
#9=IGNORADO
```

## Sintomas comuns e evolução do paciente

#### 1=Cura
#### 2=Óbito
#### 9= Ignorado

```python
df.query('ID_MUNICIP=="SAO JOSE DOS CAMPOS"')[['EVOLUCAO','NU_IDADE_N','CS_SEXO','DIARREIA','VOMITO','ASMA','DIABETES','GARGANTA','TOSSE','FEBRE']].head(40).round()
```

### Casos de SRAG em São José dos Campos

#### 1=Cura
#### 2=Óbito
#### 9= Ignorado

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('EVOLUCAO').size().plot(kind='bar',figsize=(9,9))
```

## Quantide de Pessoas que viajaram para fora do Brasil 14 dias antes dos sintomas
#### Por estado

```python
df.query('HISTO_VGM == 1 and PAIS_VGM != "BRASIL"').groupby('SG_UF').size().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(7,7))
```

## Número de casos por estado

```python
df.query('CLASSI_FIN== 5').groupby('SG_UF').size().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(7,7))
```
